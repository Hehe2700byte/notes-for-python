```python
#filter by conditions
# Create a new DataFrame of only tall Pokemon
tall_pokemon = df[df['Height'] >= 2]

# Using logical operators (OR) to find Water-type Pokemon in either Slot 1 or Slot 2
water_pokemon = df[(df['Type1'] == 'Water') | (df['Type2'] == 'Water')]
```

```python
#analyzing data with aggretgate functions
average_weight = df['Weight'].mean()
print(f"Average weight: {average_weight}")

type1_group = df.groupby('Type1')
print("Average height of groups divided by Type1")
print(type1_group['Height'].mean())
```

Data Cleaning:
Since roughly 75% of pandas work is data cleaning, you must handle missing values or inconsistent formatting

Handling Missing Data: Many Pokémon lack a second type, resulting in NaN (Not a Number) values. You can fill these with a placeholder like "None"

Standardizing Text: You can make all names lowercase for consistency

Changing Types: Convert numeric flags (0 or 1) into booleans (True or False)

```python
#Handling missing data
'''Entities without type2 will be filled with None'''
df['Type2'] = df['Type2'].fillna("None")

#Standardizing text:
'''This won't modify the original data,
but if you extract data from "df", you will find that it has been modified.'''
df.index = df.index.str.lower()#"Name" has already become a index, it is not a column for compiler

df['Type1'] = df['Type1'].str.lower()
df['Type2'] = df['Type2'].str.lower()

print("Modified Name: ")
print(df.index)
```

```python
#Changing types
#0, 1 -> False, True
df['Legendary'] = df['Legendary'].astype(bool)
```

```python
#The final data
print(df.to_string())
```
